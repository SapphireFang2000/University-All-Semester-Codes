# Assignment_Question-1

```python
# Function to create SparkSession
def create_session():
    return SparkSession.builder \
        .appName("WordFrequency") \
        .getOrCreate()


# Function to read file and create word frequency vector
def word_frequency(spark, file_path):
    # Read the file
    lines = spark.sparkContext.textFile(file_path)

    # Split each line into words and count the occurrences of each word
    # Using a space as a delimiter explicitly
    word_counts = lines.flatMap(lambda line: line.split(" ")) \
                       .map(lambda word: (word, 1)) \
                       .reduceByKey(lambda a, b: a + b)

    return word_counts
# Common words to filter out
common_words = ["us", "has", "all", "they", "from", "who", "what", "on", "by", "more", 
                "as", "not", "their", "can", "new", "it", "but", "be", "are", "--", 
                "i", "have", "this", "will", "for", "with", "is", "that", "in", "our", 
                "we", "a", "of", "to", "and", "the", "that's", "or", "make", "do", 
                "you", "at", "it's", "than", "if", "know", "last", "about", "no", "just", 
                "now", "an", "because", "<p>we", "why", "we'll", "how", "two", "also", "every", 
                "come", "we've", "year", "over", "get", "take", "one", "them", "we're", "need", 
                "want", "when", "like", "most", "-", "been", "first", "where", "so", "these", 
                "they're", "good", "would", "there", "should", "-->", "<!--", "up", "i'm", "his", 
                "their", "which", "may", "were", "such", "some", "those", "was", "here", "she", 
                "he", "its", "her", "his", "don't", "i've", "what's", "didn't", "shouldn't", "(applause.)", 
                "let's", "doesn't"]

if __name__ == "__main__":
    # Create SparkSession
    spark = create_session()

    # Path to the file
    file_path = "/content/drive/MyDrive/Project/2010 Movies Plot.txt"

    # Create word frequency vector
    word_counts = word_frequency(spark, file_path)

word_counts.collect()
```

# Assignment_Question-2

```python
# Filter out common words
filtered_word_counts = word_counts.filter(lambda word_count: word_count[0] not in common_words)

filtered_word_counts.collect()
```

## Assignmet_Question-3

```python
# Sort the word frequency vector by frequency in descending order
sorted_word_freq_vector = filtered_word_counts.sortBy(lambda pair: pair[1], ascending=False)

sorted_word_freq_vector.collect()
```
